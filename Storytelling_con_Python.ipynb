# Storytelling

Es la técnica y/o habilidad de ***contar historias*** que conecten emocionalmente con una audiencia.

Estas pueden:


*   Transmitir ideas.
*   Vender productos.
*   Explicar datos.
*   Inspirar un cambio.

Sin embargo, queda comprobado que no se trata únicamente de "contar historias", si no, de darle sentido y propósito.

# ¿Por qué funciona?

Porque el cerebro humano tiene la capacidad de recordar historias y no datos sueltos. Es decir, una gráfica sin contexto dice poco; en cambio, una gráfica que cuenta que día de la semana se tuvo mejor venta, por la promoción nueva, se graba en la memoria y a veces, también en el corazón.


```python
# Ejemplo sencillo, con datos manuales inventados,
#para dar a entender la idea central de lo que es "Storytelling"


import pandas as pd
import matplotlib.pyplot as plt

# Datos simulados
ventas = pd.DataFrame({
    'Día': ['Lunes', 'Martes', 'Miércoles', 'Jueves', 'Viernes'],
    'Monto': [200, 450, 300, 500, 1000]
})

# Storytelling: queremos mostrar cómo la venta sube en viernes.
plt.figure(figsize=(8,5))
plt.bar(ventas['Día'], ventas['Monto'], color='khaki')
plt.title('¡El viernes es el mejor día en ventas!')
plt.xlabel('Día de la semana')
plt.ylabel('Monto vendido ($)')
plt.show()
```
